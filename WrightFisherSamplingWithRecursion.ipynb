```python
# NOTE: python has a built in recursion limit of 1000 to prevent recursion from
# crashing the program on our little baby computers
# there is a simple way to tell python to ignore or change this recursion limit
# if we choose to (although i think we should use the iterative program once 
# we use the recursive one to confirm its accuracy 


# WrightFisherSamplingWithRecursion.ipynb

# citing my sources: code itself is written by gemini using the design of a
# program written by mia, and edited by mia
# aka machine coded, human programmed


import numpy as np

"""
Purpose statement: Modeling random genetic drift using Wright-Fisher sampling.
Specifically, we will take an input of the number of diploid individuals,
the initial allele count, and the generation time (which always starts as being
equal to zero the first time the recursive function is called) and outputs
(1) a boolean representing whether the allele or its competitor is the only one
remaining and an int showing at which generation this homogeneity occurred and
(2) a list of lists of ints that tracks the history of allele frequencies at
every generation, which can later be used to create a graphical representation
Signature: int + int + int => bool + int + list(list(int))
I/O spec
Input:
N = int, the number of diploid individuals
K = int, the allele count
t = int, generation time
Output:
b = bool, true if allele has won natural selection, false if allele has lost
natural selection (allele fixation)
t = int, generation time
history = list(list(int)) showing the history of k, allele count
(note that k/2N = p) at every generation t. In a separate function we can use
history to create a graphical representation of the change in allele frequency
over time

Recursion diagrams:
note: because it's probabilistic, the likelihood of these recursion diagrams
going exactly like this if we were to run the code is quite slim if not zero,
the goal here is just to understand how the program might operate

OI: N = 50, K = 10 , t = 0
RI: N = 50, K = 5, t = 1
RO: b = false, t = 2, history = [ [ 1, 5] , [2, 0] ]
OO: b = false, t = 2, history = [ [0 , 10] , [ 1, 5] , [2, 0] ]

OI: N = 100, K = 140 , t = 0
RI: N = 100, K = 164 , t = 1
RO: b = true, t = 3, history =  [ [1, 164], [2, 190], [3, 200] ]
OO: b = true, t = 3, history = [ [0, 140] , [1, 164], [2, 190], [3, 200] ]

"""

def wright_fisher_sampling(N: int, K: int, t: int, history: list):


    # Input validation
    if N <= 0:
        raise ValueError("population number must be greater than zero.")
    if K < 0:
        raise ValueError("number of alleles cannot be negative.")
    if K > 2 * N:
        raise ValueError("K cannot be greater than 2*N (total alleles in pop)")

    if history is None:
        history = []
    history.append([t, K])

    # Base cases: Allele fixation or loss
    if K == 0:
        return False, t, history
    if K == 2 * N:
        return True, t, history

    # Calculate allele frequency
    p = K / (2 * N)

    # Sample K' for the next generation using binomial distribution
    # 2*N trials (total alleles), probability p of success (allele 'A')
    next_K = np.random.binomial(2 * N, p)

    # Recursive call for the next generation
    return wright_fisher_sampling(N, next_K, t + 1, history)

# --- Test Cases

print("Input: N = 50, K = 10 , t = 0")
print(wright_fisher_sampling(50, 10, 0, None))

print("\nInput: N = 100, K = 70 , t = 0")
print(wright_fisher_sampling(100, 70, 0, None))

print("\nInput: N = 100, K = 0, t = 0")
print(wright_fisher_sampling(100, 0, 0, None))

print("\nInput: N = 100, K = 100, t = 0")
print(wright_fisher_sampling(100, 100, 0, None))

print("\nInput: N = 100, K = 99, t = 0")
print(wright_fisher_sampling(100, 99, 0, None))

print("\nInput: N = 100, K = 3, t = 0")
print(wright_fisher_sampling(100, 3, 0, None))

print("\nInput: N = 1, K = 1, t = 0")
print(wright_fisher_sampling(1, 1, 0, None))

print("\nInput: N = 1, K = 2, t = 0")
print(wright_fisher_sampling(1, 2, 0, None))

print("\nInput: N = 100, K = 60, t = 68")
print(wright_fisher_sampling(100, 60, 68, None))

# Edge cases and error handling
print("\nInput: N = 100, K = -2, t = 0")
try:
    wright_fisher_sampling(100, -2, 0, None)
except ValueError as e:
    print(f"Expected Error: {e}")

print("\nInput: N = 0, K = 0, t = 0")
try:
    wright_fisher_sampling(0, 0, 0, None)
except ValueError as e:
    print(f"Expected Error: {e}")

print("\nInput: N = -60, K = 39, t = 0")
try:
    wright_fisher_sampling(-60, 39, 0, None)
except ValueError as e:
    print(f"Expected Error: {e}")

print("\nInput: N = 20, K = 50, t = 0") # K > 2*N condition
try:
    wright_fisher_sampling(20, 50, 0, None)
except ValueError as e:
    print(f"Expected Error: {e}")

```
